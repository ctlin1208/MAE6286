```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
#paremeters
nx = 41
L = 2.0
dx = L/(nx - 1)
nt = 25
dt = 0.02
c = 1.0

#define the grid point coodinates
x = numpy.linspace(0.0, L, num=nx)
```

```python
# initial conditions with 1.0 everywhere (for now)
u0 = numpy.ones(nx)

# get a list of indices where 0.5 <= x <= 1.0
mask = numpy.where(numpy.logical_and(x >= 0.5, x <= 1.0))
print(mask)
```

```python
#initial condition u = 2.0 where 0.5 <= x <=1.0
u0[mask] = 2.0
print(u0)
```

```python
#plot the initial conditions
pyplot.figure(figsize=(4.0, 4.0))
pyplot.title('initial conditions')
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0, linestyle='--')
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```

```python
u = u0.copy()
for n in range(1, nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - c*dt/dx*(un[i] - un[i - 1])
```

```python
#plot the solution after nt time steps
#along with the initial conditions
pyplot.figure(figsize=(4.0, 4.0))
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0, label='Initial', linestyle='--')
pyplot.plot(x, u, label='nt = {}'.format(nt))
pyplot.legend()
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```

```python
#paremeters
nx = 41
L = 2.0
dx = L/(nx - 1)
nt = 10
dt = 0.02

x = numpy.linspace(0.0, L, num=nx)
u0 = numpy.ones(nx)
mask = numpy.where(numpy.logical_and(x >= 0.5, x <= 1.0))
u0[mask] = 2.0
```

```python
#plot the initial conditions
pyplot.figure(figsize=(4.0, 4.0))
pyplot.title('initial conditions')
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0, linestyle='--')
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```

```python
# computethe solution using Euler's method and array slicing
u = u0.copy()
for n in range(1, nt):
    u[1:] = u[1:] - dt/dx*u[1:]*(u[1:] - u[:-1])
```

```python
#plot the solution after nt time steps
#along with the initial conditions
pyplot.figure(figsize=(4.0, 4.0))
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u0, label='Initial', linestyle='--')
pyplot.plot(x, u, label='nt = {}'.format(nt))
pyplot.legend()
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```
