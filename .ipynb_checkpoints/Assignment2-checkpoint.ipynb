```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
#Set the initial conditions
Vmax = 80
L = 22
rhomax = 250
nx = 101
dt = 0.001
dx = L / (nx - 1)
nt = 25

x = numpy.linspace(0.0, L, num=nx)
rho0 = numpy.ones(nx)
mask = numpy.where(numpy.logical_and(x >= 0.5, x <= 1.0))
rho0[mask] = 2.0
```

```python
rho = rho0.copy()
for n in range(1, nt):
    rhon = rho.copy()
    for i in range(1, nx):
        rho[i] = rhon[i] + Vmax*(rho[i]/rhomax - 1) * dt / dx * (rhon[i] - rhon[i - 1])
```

```python
# Plot the solution after nt time steps
# along with the initial conditions.
pyplot.figure(figsize=(4.0, 4.0))
pyplot.xlabel('x')
pyplot.ylabel('rho')
pyplot.grid()
pyplot.plot(x, rho0, label='Initial',
            color='C0', linestyle='--', linewidth=2)
pyplot.plot(x, rho, label='nt = {}'.format(nt),
            color='C1', linestyle='-', linewidth=2)
pyplot.legend()
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```

```python
def u0(x):
    if 2 <= x <= 4.2:
        return 50
    else:
        return 10

pyplot.figure(figsize=(4.0, 4.0))
pyplot.xlabel('x')
pyplot.ylabel('rho0')
pyplot.grid()
pyplot.plot(x, u0)
pyplot.legend()
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 100);
```
